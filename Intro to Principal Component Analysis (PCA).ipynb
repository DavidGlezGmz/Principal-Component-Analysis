```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
from numpy.linalg import eig

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
n = 1000
dims = 2
X = np.random.randn(n, dims)@np.array([[2,1],[1,2]])
```

```python
plt.scatter(X[:,0], X[:,1], s=8)
```

```python
eig(X.T@X)
```

```python
eigen_values, eigen_vectors = eig(X.T@X)
```

```python
X_new = X@eigen_vectors
```

```python
plt.scatter(X_new[:,0], X_new[:,1], s=8)
```

```python
def PCA(X,dims):
  eigen_values, eigen_vectors = eig(X.T@X)
  sort_indices = np.argsort(eigen_values)
  eigen_vectors= eigen_vectors[:, sort_indices[:dims]]
  return X@eigen_vectors
```

```python
X_pca = PCA(X, 2)
plt.scatter(X_pca[:,0], X_pca[:,1], s=8)
```

```python
n = 1000
dims = 3
X = np.random.randn(n, dims)@np.array([[1,1,1],[1,1,1],[2,1,1]])
fig = plt.figure()
ax_3d = fig.add_subplot(projection='3d')
ax_3d.scatter(X[:,0], X[:,1], X[:,2])
```

```python
X_pca2=PCA(X,2)
plt.scatter(X_pca2[:,0], X_pca2[:,1], s=4)
```

```python
X_pca3 = PCA(X,3)
```

```python
fig2 = plt.figure()
ax_3d = fig2.add_subplot(projection='3d')
ax_3d.scatter(X_pca3[:,0], X_pca3[:,1], X_pca3[:,2])
```

```python
X_donut=np.random.randn(10000,3)*5
dist=np.sqrt(np.sum(X_donut**2, axis=1))
X_donut=X_donut[dist>10]
y_donut=np.ones([X_donut.shape[0],1])

X_hole=np.random.randn(1000,3)
y_hole=np.zeros([X_hole.shape[0],1])

X=np.vstack([X_donut,X_hole])
y=np.vstack([y_donut,y_hole])
```

```python
fig4 = plt.figure()
ax_3d = fig4.add_subplot(projection='3d')
ax_3d.scatter(X[:,0], X[:,1], X[:,2], s=2, c=y.flatten(), cmap='coolwarm')
```
